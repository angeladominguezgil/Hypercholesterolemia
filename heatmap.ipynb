```python
#Import functions as needed
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
#Load excel file 
file="heatmap for python.xlsx"
df=pd.read_excel(file, sheet_name='map',engine='openpyxl')
df.head(74)
```

```python
#Set data information for the heatmap
#df.iloc[:74] #From line 74 there is no significant data
df.set_index('Drug name', inplace=True)
prot_data=df.iloc[:74,2:]
group_data=df['Group'].tolist()
```

```python
#Heatmap plotting

plt.figure(figsize=(10,20))
graph=sns.heatmap(prot_data, cmap=sns.color_palette("RdYlGn", as_cmap=True),
            annot=False, vmin=0.15,vmax=0.95, cbar_kws={'shrink': 0.4,'location':'left'}, cbar=True)
#plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=90, fontsize=6)
plt.yticks(rotation=0,fontsize=6)
plt.tight_layout()


current_group = group_data[0]
group_start = 0
for i in range(1, len(group_data)):
    if group_data[i] != current_group:
        graph.axhline(i, color='black', linewidth=1)
        plt.text(prot_data.shape[1] + 0.5, (group_start + i) / 2, current_group,verticalalignment='center', fontsize=14, rotation=0,bbox=dict(facecolor='lightblue', edgecolor='none', boxstyle='round,pad=0.3'))
        current_group = group_data[i]
        group_start = i



plt.show()
```
